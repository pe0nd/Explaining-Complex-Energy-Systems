# Moral Graph

In graph theory, a moral graph is used to find the equivalent undirected form of a directed acyclic graph. It is a key
step of the junction tree algorithm, used in belief propagation on graphical models.

The moralized counterpart of a directed acyclic graph is formed by adding edges between all pairs of non-adjacent nodes
that have a common child, and then making all edges in the graph undirected. Equivalently,a moral graph of a directed
acyclic graph G is an undirected graph in which each node of the original G is now connected to its Markov blanket. The
name stems from the fact that, in a moral graph, two nodes that have a common child are required to be married by
sharing an edge.

```python
import igraph as ig

g1 = ig.Graph(directed=True)
g1.add_vertices(["1","2","3","4","5","6","7"])
g1.add_edges([(0,2),(1,2),(1,3),(2,4),(2,5),(2,6),(3,6)])
g1.vs["label"]=g1.vs["name"]
ig.plot(g1, layout=g1.layout("rt"), bbox=(300,150), margin=20)
```

```python
import igraph as ig

g1 = ig.Graph(directed=False)
g1.add_vertices(["1","2","3","4","5","6","7"])
g1.add_edges([(0,2),(1,2),(1,3),(2,4),(2,5),(2,6),(3,6),(0,1),(2,3)])
g1.vs["label"]=g1.vs["name"]
g1.es.select(8,7)["color"]= "blue"

ig.plot(g1, layout=g1.layout("kk"), bbox=(300,150), margin=20)
```
